### 1. Text Preprocessing (í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬)
ì„¤ëª…) ëª¨ë¸ì´ í…ìŠ¤íŠ¸ë¥¼ ì´í•´í•˜ê¸° ì „ì—, ì‚¬ëŒì´ ì‚¬ìš©í•˜ëŠ” ì–¸ì–´ë¥¼ ê¸°ê³„ê°€ ì²˜ë¦¬í•˜ê¸° ì¢‹ì€ í˜•íƒœë¡œ ë‹¤ë“¬ëŠ” ê³¼ì •ì´ í•„ìš”í•©ë‹ˆë‹¤. ì´ë¥¼ í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬ë¼ê³  í•©ë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” ê°€ì¥ ê¸°ë³¸ì ì¸ **í† í°í™”(Tokenization)**ì™€ **ë¶ˆìš©ì–´ ì œê±°(Stopword Removal)**ë¥¼ ì‹¤ìŠµí•´ ë³´ê² ìŠµë‹ˆë‹¤.

```python
# nltk ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜
!pip install nltk

import nltk
from nltk.tokenize import word_tokenize

# ë°ì´í„° ë‹¤ìš´ë¡œë“œ
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# í…ìŠ¤íŠ¸
text = "ìì—°ì–´ ì²˜ë¦¬, ì¦‰ NLPëŠ” ë§¤ìš° í¥ë¯¸ë¡œìš´ ì—°êµ¬ ë¶„ì•¼ì…ë‹ˆë‹¤. ì»´í“¨í„°ê°€ ì¸ê°„ì˜ ì–¸ì–´ë¥¼ ì´í•´í•˜ê³  ì²˜ë¦¬í•  ìˆ˜ ìˆë„ë¡ í•´ì¤ë‹ˆë‹¤."

# 1. í† í°í™”:
tokens = word_tokenize(text)
print("1. ì›ë³¸ í† í°:", tokens)

# 2. ë¶ˆìš©ì–´ ì œê±°:
stop_words = ['ì¦‰', 'ëŠ”', 'ê°€', 'ì´', 'ì„', 'ë¥¼', 'ì˜', 'ì—', 'ë“¤', 'ì€', 'ì…ë‹ˆë‹¤', 'í•˜ê³ ', 'ìˆ˜', 'ìˆë„ë¡']
filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

print("\n2. ì§ì ‘ ì •ì˜í•œ ë¶ˆìš©ì–´:", sorted(list(stop_words)))
print("  ê¸°ì¡´ ë¶ˆìš©ì–´ ì œê±° í›„ í† í°:", filtered_tokens)

# ----- ë¬¸ì œ1) stop_words ë¦¬ìŠ¤íŠ¸ ë¹ˆì¹¸ ì±„ìš°ì„¸ìš”.
stop_words.extend(['ë¶„ì•¼', 'ì—°êµ¬'])

# ì¶”ê°€ëœ ë¶ˆìš©ì–´ ëª©ë¡ì„ ì ìš©í•˜ì—¬ ë‹¤ì‹œ í•„í„°ë§
problem1_filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

print("'ë¶„ì•¼', 'ì—°êµ¬' ì¶”ê°€ í›„ ê²°ê³¼:", problem1_filtered_tokens)
```

#### âœ… ë¬¸ì œ1) ìœ„ ì½”ë“œì˜ stop_wordsì— **'ë¶„ì•¼', 'ì—°êµ¬'**ë¥¼ ì¶”ê°€í•˜ì—¬ ë¶ˆìš©ì–´ë¡œ ì²˜ë¦¬í•œ í›„, ê²°ê³¼ê°€ ì–´ë–»ê²Œ ë‹¬ë¼ì§€ëŠ”ì§€ í™•ì¸í•´ë³´ì„¸ìš”. (ê¸°ì¡´ ê²°ê³¼ì™€ ë¶ˆìš©ì–´ ì²˜ë¦¬ í›„ ê²°ê³¼ì™€ ë¹„êµí•´ ì§„í–‰í•´ë³´ê¸°)

#### âœ… ë¬¸ì œ2) í† í°í™” ê²°ê³¼ì— í¬í•¨ëœ ','ë‚˜ '.' ê°™ì€ êµ¬ë‘ì ë“¤ì´ ë¶ˆìš©ì–´ ì œê±° ê³¼ì •ì—ì„œ ì–´ë–»ê²Œ ì²˜ë¦¬ë˜ì—ˆë‚˜ìš”? ê·¸ ì´ìœ ëŠ” ë¬´ì—‡ì¼ê¹Œìš”?

- ì‚¬ë¼ì¡Œë‹¤. "word for word in tokens if word.isalnum()" ì½”ë“œì˜ word.isalnum()ì€ ì•ŒíŒŒë²³ê³¼ ìˆ«ìë§Œ í—ˆìš©í•˜ê³  êµ¬ë‘ì ì´ë‚˜ íŠ¹ìˆ˜ë¬¸ìëŠ” ìë™ìœ¼ë¡œ ì œì™¸í•˜ëŠ” í•¨ìˆ˜ì´ê¸° ë•Œë¬¸ì´ë‹¤.

### 2. N-gram
ì„¤ëª…) N-gramì€ í…ìŠ¤íŠ¸ì—ì„œ ì—°ì†ëœ nê°œì˜ ë‹¨ì–´ ë­‰ì¹˜ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤. N-gramì€ ë¬¸ë§¥ì„ íŒŒì•…í•˜ëŠ” ê°„ë‹¨í•˜ë©´ì„œë„ íš¨ê³¼ì ì¸ ë°©ë²•ì…ë‹ˆë‹¤. 1-gramì€ ìœ ë‹ˆê·¸ë¨(unigram), 2-gramì€ ë°”ì´ê·¸ë¨(bigram), 3-gramì€ íŠ¸ë¼ì´ê·¸ë¨(trigram)ì´ë¼ê³ ë„ ë¶€ë¦…ë‹ˆë‹¤.

```python
def generate_ngrams(text, n):
    words = text.split()
    ngrams = zip(*[words[i:] for i in range(n)])
    return [" ".join(ngram) for ngram in ngrams]

# ì‹œì‘ ë¬¸ì¥ (ìˆ˜ì • ëŒ€ìƒ)
sentences = [
  "ì˜¤ëŠ˜ ë‚ ì”¨ê°€ ì •ë§ ì¢‹ë„¤ìš”.",
  "ì €ëŠ” ì•„ì¹¨ì— ì»¤í”¼ë¥¼ ë§ˆì‹­ë‹ˆë‹¤.",
  "ì£¼ë§ì— ì¹œêµ¬ë¥¼ ë§Œë‚˜ê¸°ë¡œ í–ˆì–´ìš”.",
  "í‡´ê·¼í•˜ê³  ì§‘ì— ê°€ëŠ” ê¸¸ì…ë‹ˆë‹¤.",
  "ì˜¤ëŠ˜ ì €ë… ë©”ë‰´ëŠ” ë¬´ì—‡ì¸ê°€ìš”?",
  "ìŒì•… ë“¤ìœ¼ë©° ì‚°ì±…í•˜ëŠ” ê²ƒì„ ì¢‹ì•„í•©ë‹ˆë‹¤.",
  "ë³´ê³  ì‹¶ì€ ì˜í™”ê°€ ìƒê²¼ì–´ìš”.",
  "ê³§ ê°œê°•ì´ë¼ ì•„ì¹¨ ì¼ì° ì¼ì–´ë‚˜ì•¼ í•©ë‹ˆë‹¤."
]

# ì´ˆê¸° N-gram ê°œìˆ˜ í™•ì¸
initial_bigrams = sum(len(generate_ngrams(s, 2)) for s in sentences)
initial_trigrams = sum(len(generate_ngrams(s, 3)) for s in sentences)

print(f"Initial Bigram Count: {initial_bigrams}")   # ê¸°ì¡´ ê²°ê³¼: 15
print(f"Initial Trigram Count: {initial_trigrams}") # ê¸°ì¡´ ê²°ê³¼: 10
```

#### âœ… ë¬¸ì œ1 ) ì•„ë˜ sentences ë°°ì—´ì˜ ë¬¸ì¥ë“¤ì„ ìˆ˜ì •í•˜ê±°ë‚˜ ìƒˆë¡œìš´ ë¬¸ì¥ì„ ì¶”ê°€í•˜ì—¬, ì „ì²´ 2-gramì˜ ìˆ˜ê°€ 25ê°œ ì´ìƒ, 3-gramì˜ ìˆ˜ê°€ 15ê°œ ì´ìƒì´ ë˜ë„ë¡ ë§Œë“¤ì–´ë³´ì„¸ìš”.

#### âœ… ë¬¸ì œ2) N-gram ëª¨ë¸ì€ ê°„ë‹¨í•˜ì§€ë§Œ í•œê³„ê°€ ìˆìŠµë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´, "ê°•ì•„ì§€ê°€ ê³ ì–‘ì´ë¥¼ ì¢‹ì•„í•´"ì™€ "ê³ ì–‘ì´ê°€ ê°•ì•„ì§€ë¥¼ ì¢‹ì•„í•´"ëŠ” 1-gramì˜ êµ¬ì„±ê³¼ ë¹ˆë„ëŠ” ê°™ì§€ë§Œ ì˜ë¯¸ëŠ” ë‹¤ë¦…ë‹ˆë‹¤. ì´ì²˜ëŸ¼ N-gramì´ ë¬¸ì¥ì˜ êµ¬ì¡°ì , ì˜ë¯¸ì  ì •ë³´ë¥¼ ì™„ì „íˆ íŒŒì•…í•˜ê¸° ì–´ë ¤ìš´ ì´ìœ ëŠ” ë¬´ì—‡ì¼ê¹Œìš”?

- N-gram ëª¨ë¸ì€ ë¬¸ì¥ì„ ì—°ì†ëœ í† í° ë¬¶ìŒìœ¼ë¡œ ë³´ëŠ” í•œê³„ê°€ ìˆê¸° ë•Œë¬¸ì´ë‹¤.

### 5. Sequence Models (RNNì˜ í•œê³„)
ì„¤ëª… ) RNN(ìˆœí™˜ ì‹ ê²½ë§)ì€ ì‹œí€€ìŠ¤ ë°ì´í„° ì²˜ë¦¬ì— ê°•ì ì„ ê°–ì§€ë§Œ, ì‹œí€€ìŠ¤ê°€ ê¸¸ì–´ì§ˆìˆ˜ë¡ ì•ìª½ì˜ ì •ë³´ê°€ ë’¤ìª½ìœ¼ë¡œ ì „ë‹¬ë˜ê¸° ì–´ë ¤ìš´ **ì¥ê¸° ì˜ì¡´ì„± ë¬¸ì œ(Long-term Dependency Problem)**ë¼ëŠ” êµ¬ì¡°ì  í•œê³„ë¥¼ ê°€ì§‘ë‹ˆë‹¤. ì´ëŠ” Transformerì™€ Attentionì´ ë“±ì¥í•˜ê²Œ ëœ ì¤‘ìš”í•œ ë°°ê²½ì´ ë©ë‹ˆë‹¤.

ì´ ë¬¸ì¥ì—ì„œ ë§ˆì§€ë§‰ ë‹¨ì–´ì¸ 'ë‹´ë‹¹ìì´ë‹¤'ëŠ” ë¬¸ì¥ ë§¨ ì•ì˜ 'ê¹€ ëŒ€ë¦¬'ì™€ ê°€ì¥ ë°€ì ‘í•œ ê´€ë ¨ì´ ìˆìŠµë‹ˆë‹¤. RNNì´ ì´ì²˜ëŸ¼ ë©€ë¦¬ ë–¨ì–´ì§„ ë‹¨ì–´ ê°„ì˜ ê´€ê³„ë¥¼ íŒŒì•…í•˜ê¸° ì–´ë ¤ìš´ ì´ìœ ëŠ” ë¬´ì—‡ì¼ê¹Œìš”? (Vanishing Gradient(ê¸°ìš¸ê¸° ì†Œì‹¤) ê´€ì ì—ì„œ ê°„ë‹¨íˆ ì„¤ëª…í•´ë³´ì„¸ìš”.) <br>

### 6. Attention, Transformer

1. . Transformer êµ¬ì¡°
Query, Key, Value ì†Œê°œ

Attention ë©”ì»¤ë‹ˆì¦˜ì€ Query(ì¿¼ë¦¬), Key(í‚¤), Value(ê°’) ë¼ëŠ” ì„¸ ê°€ì§€ ê°œë…ì„ ì‚¬ìš©í•´ ë¬¸ì¥ ë‚´ ë‹¨ì–´ ê°„ì˜ ê´€ê³„ ê°€ì¤‘ì¹˜ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤. ë„ì„œê´€ì—ì„œ ì±…ì„ ì°¾ëŠ” ê²ƒì— ë¹„ìœ í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

Query: ë‚´ê°€ ì°¾ê³  ì‹¶ì€ ì •ë³´

Key: ê° ì±…ì˜ ì£¼ì œë‚˜ ì œëª©

Value: ì±…ì˜ ì‹¤ì œ ë‚´ìš©

Attentionì€ ë‚´ Queryì™€ ê°€ì¥ ê´€ë ¨ì„±ì´ ë†’ì€ Key(ìœ ì‚¬ë„ ê³„ì‚°)ë¥¼ ì°¾ê³ , í•´ë‹¹ Keyì— ì—°ê²°ëœ Value(ì‹¤ì œ ë‚´ìš©)ë¥¼ ë” ë§ì´ ê°€ì ¸ì˜¤ëŠ” ë°©ì‹ìœ¼ë¡œ ë™ì‘í•©ë‹ˆë‹¤. Self-Attentionì—ì„œëŠ” ì´ ì„¸ ê°€ì§€ê°€ ëª¨ë‘ ê°™ì€ ë¬¸ì¥ ë‚´ ë‹¨ì–´ë“¤ë¡œë¶€í„° ë‚˜ì˜µë‹ˆë‹¤.

ğŸ’» Self-Attention ì‹¤ìŠµ

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x, axis=-1):
    exp_x = np.exp(x - np.max(x, axis=axis, keepdims=True))
    return exp_x / np.sum(exp_x, axis=axis, keepdims=True)

# ë¬¸ì¥: "ê·¸ ë™ë¬¼ì€ ê¸¸ì„ ê±´ë„ˆì§€ ì•Šì•˜ë‹¤. ì™œëƒí•˜ë©´ ê·¸ê²ƒ(it)ì´ ë„ˆë¬´ í”¼ê³¤í–ˆê¸° ë•Œë¬¸ì´ë‹¤."
tokens = "The animal didn't cross the street because it was too tired".split()

# ì„ì˜ì˜ ì„ë² ë”© ìƒì„± (ì‹¤ì œë¡œëŠ” í•™ìŠµëœ ì„ë² ë”© ì‚¬ìš©)
embedding_dim = 64
np.random.seed(42)
X = np.random.rand(len(tokens), embedding_dim)

Q, K, V = X, X, X
d_k = K.shape[-1]

# Attention ì ìˆ˜ ê³„ì‚°
scores = np.dot(Q, K.T) / np.sqrt(d_k)
attention_weights = softmax(scores)

# 'it'ì´ Queryì¼ ë•Œì˜ Attention ê°€ì¤‘ì¹˜ ì¶œë ¥
it_index = tokens.index('it')
it_attention = attention_weights[it_index]

print("Attention weights for the query 'it':")
for token, weight in zip(tokens, it_attention):
    print(f"{token}: {weight:.4f}")

# Heatmap ì‹œê°í™”
plt.figure(figsize=(7, 6))
plt.imshow(attention_weights, cmap='viridis')
plt.colorbar(label='Attention Weight')
plt.xticks(ticks=range(len(tokens)), labels=tokens, rotation=45)
plt.yticks(ticks=range(len(tokens)), labels=tokens)
plt.title("Self-Attention Heatmap")
plt.xlabel("Key Tokens")
plt.ylabel("Query Tokens")
plt.tight_layout()
plt.show()
```

#### âœ… ë¬¸ì œ1) ìœ„ ì½”ë“œì˜ ì‹¤í–‰ ê²°ê³¼(Attention ê°€ì¤‘ì¹˜)ì™€ íˆíŠ¸ë§µì„ í™•ì¸í•˜ì„¸ìš”. Queryê°€ 'it'ì¼ ë•Œ, 'animal'ê³¼ 'street' ì¤‘ ì–´ë–¤ Keyì˜ ê°€ì¤‘ì¹˜(Attention Score)ê°€ ë” ë†’ê²Œ ë‚˜ì™”ë‚˜ìš”? (ì‹¤í–‰ ì‹œë§ˆë‹¤ ê²°ê³¼ëŠ” ë‹¤ë¥¼ ìˆ˜ ìˆì§€ë§Œ ì¼ë°˜ì ì¸ ê²½í–¥ì„±ì„ ë³´ì„¸ìš”.)<br>
- animalì˜ ê°€ì¤‘ì¹˜ê°€ ë” ë†’ê²Œ ë‚˜ì™”ë‹¤.

#### âœ… ë¬¸ì œ2) Self-Attentionì€ ì™œ RNNì˜ ì¥ê¸° ì˜ì¡´ì„± ë¬¸ì œë¥¼ í•´ê²°í•  ìˆ˜ ìˆì„ê¹Œìš”? (ê³„ì‚° ë°©ì‹ì˜ ì°¨ì´ì ì„ ìƒê°í•´ë³´ì„¸ìš”.) <br>
- RNNì€ "it"ì´ "animal"ì„ ì°¸ì¡°í•  ë•Œ ì—¬ëŸ¬ ì€ë‹‰ ìƒíƒœë¥¼ ê±°ì¹˜ë©° ì—­ì „íŒŒ ê³¼ì •ì—ì„œ ê¸°ìš¸ê¸°ê°€ 0ì— ê°€ê¹Œì›Œì§„ë‹¤. ê·¸ëŸ¬ë‚˜ self-Attentionì€ ì…ë ¥ ë¬¸ì¥ ì „ì²´ë¥¼ í•œ ë²ˆì— ë°›ì•„ Queryì™€ Key ì‚¬ì´ì˜ ê°€ì¤‘ì¹˜ë¥¼ ê³„ì‚°í•˜ë¯€ë¡œ "it"ì´ "animal"ì„ ì°¸ì¡°í•  ë•Œ ì¤‘ê°„ì— ê±°ì¹˜ëŠ” ë‹¨ê³„ ì—†ì´ ë°”ë¡œ ì—°ê²°ë˜ê²Œ ëœë‹¤. ë©€ë¦¬ ë–¨ì–´ì§„ ë‹¨ì–´ë¼ë„ ê°€ì¤‘ì¹˜ ê³„ì‚°ì´ ë°”ë¡œ ì´ë£¨ì–´ì§€ë¯€ë¡œ ì¥ê¸° ì˜ì¡´ì„± ë¬¸ì œë¥¼ í•´ê²°í•  ìˆ˜ ìˆë‹¤.

####ğŸ’» Masked Self-Attention (ë””ì½”ë”) ì‹¤ìŠµ
ì„¤ëª…) Transformerì˜ ë””ì½”ë”ëŠ” ë‹¤ìŒ ë‹¨ì–´ë¥¼ ì˜ˆì¸¡í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤. ì´ë•Œ, ì •ë‹µì„ ë¯¸ë¦¬ ë³´ê³  ì˜ˆì¸¡í•˜ëŠ” ê²ƒì„ ë°©ì§€í•˜ê¸° ìœ„í•´ í˜„ì¬ ì˜ˆì¸¡í•˜ë ¤ëŠ” ìœ„ì¹˜ë³´ë‹¤ ë’¤ì— ìˆëŠ” ë‹¨ì–´ë“¤ì„ ê°€ë¦¬ëŠ” ë§ˆìŠ¤í‚¹(Masking) ê³¼ì •ì´ í•„ìš”í•©ë‹ˆë‹¤.

```python
# 5x5 í¬ê¸°ì˜ ì„ì˜ì˜ Attention ì ìˆ˜ í–‰ë ¬ ìƒì„±
np.random.seed(42)
scores = np.random.rand(5, 5)
print("Original Scores:\n", np.round(scores, 2))

# ë§ˆìŠ¤í¬ ìƒì„±
mask = np.triu(np.ones((5, 5)), k=1).astype('bool')

# ë§ˆìŠ¤í‚¹ ì ìš© (ë§ˆìŠ¤í‚¹ëœ ì ìˆ˜ì— Softmaxë¥¼ ì ìš©)
scores[mask] = -np.inf
print("\nMasked Scores:\n", np.round(scores, 2))

# Softmax ì ìš© í›„ ê²°ê³¼
attention_weights = softmax(scores, axis=-1)
print("\nAttention Weights after Masking and Softmax:\n", np.round(attention_weights, 2))
```

- ì˜ˆì¸¡ ê¸°ì¤€ ì‹œì ìœ¼ë¡œë¶€í„° ë¯¸ë˜ì— ìˆëŠ” ê²ƒì€ ì°¸ì¡°í•˜ì§€ ì•Šë„ë¡ í•˜ê¸° ìœ„í•¨ì´ë‹¤.

- í•™ìƒì„ ì˜ˆì¸¡í•˜ëŠ” ì‹œì ì—ì„œ ì…ë‹ˆë‹¤ëŠ” ë¯¸ë˜ ì‹œì ì´ë‹¤. ë¯¸ë˜ ì‹œì ì˜ ë°ì´í„°ë¥¼ ì°¸ì¡°í•  ê²½ìš° ë¯¸ë˜ ì •ë‹µì„ ë³µì‚¬í•˜ëŠ” ë™ì‘ì´ ë°œìƒí•œë‹¤. ë”°ë¼ì„œ ë¯¸ë˜ ì‹œì ì˜ ìœ„ì¹˜ì— -infë¥¼ ì…ë ¥í•˜ì—¬ softmax í›„ 0ì˜ ê°’ì„ ê°–ë„ë¡ í•œë‹¤. ì´ë ‡ê²Œ ë˜ë©´ ëª¨ë¸ì€ 0ì˜ ê°’ì„ ì§€ë‹Œ ë¯¸ë˜ ì‹œì ì˜ ë°ì´í„°ëŠ” ì°¸ì¡°í•˜ì§€ ì•Šê³  í˜„ì¬ì™€ ê³¼ê±° ì‹œì ì˜ ë°ì´í„°ë§Œ ì°¸ì¡°í•˜ê²Œ ëœë‹¤.
